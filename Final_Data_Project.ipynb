```python
#library to show the file dialog bog
from tkinter import filedialog
#library import for any path
from tkinter import *
#library import for message box
from tkinter import messagebox
#library import for ask file path
from tkinter.filedialog import asksaveasfile
#library import for gui
import tkinter as tk
#library import for plotting 
import matplotlib
#library import fior pandas table display
from pandastable import Table, TableModel
#library import for plot
import matplotlib.pyplot as plt
matplotlib.use('TkAgg')
#library import for images
from matplotlib.figure import Figure
#library import for canvas
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg,NavigationToolbar2Tk)
#library import for datasheets
import pandas as pd
#library import for numerical data
import numpy as np
#library import for Json
import json
#library import for csv
import csv
#libary import for seaborn
import seaborn as sns

#GUI Root window
root = tk.Tk()
#Title for GUI
root.title("Data Analysis Interface")
#Making the frame progragate on its own is disaled
root.pack_propagate(False)
#Creating the canvas for the window
canvas1 = tk.Canvas(root, width=900, height=800, bg='lightsteelblue2', relief='raised')
#Grid display
canvas1.grid()
#label Creation for the title
label1 = tk.Label(root, text='Data Analysis Tool from CSV to Json', bg='lightsteelblue2',)
#Label font
label1.config(font=('helvetica', 20))
canvas1.create_window(400, 60, window=label1)

# Initial set the Path to home
path = './'


# Loading a CSV file
def getCSV():
    global df3
    global de3
    global dg3
    
    #Reading all the csv files to the system
    df = pd.read_csv ('airports.csv')
    de=pd.read_csv ('airport-frequencies.csv')
    dg=pd.read_csv ('runways.csv')
    #Creating the datatframe
    df3= pd.DataFrame(df)
    de3=pd.DataFrame(de)
    dg3=pd.DataFrame(dg)
    #Creating a separate window for the message box
    MsgBox = tk.messagebox.showinfo('Message', 'Files uploaded',
                                       icon='warning')
    
    
    
#Button to import CSV Files / Fonting
browseButton_CSV = tk.Button(text="      Import CSV File     ", command=getCSV, bg='green', fg='white',
                             font=('helvetica', 12, 'bold'))
canvas1.create_window(150, 130, window=browseButton_CSV)

#Function to clean the loaded CSV
def clean_csv():
    global df3
    global de3
    global dg3
    global merged_large
    global merged_small
    global merged_medium
    #Loading the csv files
    df3 = pd.read_csv("airports.csv")
    de3 = pd.read_csv("airport-frequencies.csv")
    dg3 = pd.read_csv("runways.csv")
    #Replacing the null values with frequently occuring values
    result = df3.fillna(df3.mode().iloc[0])
    df3=result
    #Removing the closed airport 
    df3.drop(df3[df3['type'] == "closed"].index, inplace = True)
    #combining and filtering large airport with GB
    filter1=df3[(df3.type == 'large_airport') & (df3.iso_country == "GB")   ]
    #combining and filtering medium airport with GB
    filter2=df3[(df3.type == 'medium_airport') & (df3.iso_country == "GB")   ]
    #combining and filtering small airport with GB
    filter3=df3[(df3.type == 'small_airport') & (df3.iso_country == "GB")   ]
   # Combination of all the table
    de3["ident"]=de3["airport_ident"]
    dg3["ident"]=dg3["airport_ident"]
    # Removing and filtering closed airport value is 1
    dg3.drop(dg3[dg3['closed'] == 1].index, inplace = True)
    #Merging the airports and creating 3 airport table
    #Large Airport
    merged_large= pd.merge(de3,filter1, on="ident", how="inner")
    merged_large=pd.merge(dg3, merged_large ,on="ident", how="inner")
    merged_large=pd.DataFrame(merged_large)
    #Medium Airport
    merged_medium= pd.merge(de3,filter2, on="ident", how="inner")
    merged_medium=pd.merge(dg3, merged_medium ,on="ident", how="inner")
    #Small Airport
    merged_small= pd.merge(de3,filter3, on="ident", how="inner")
    merged_small=pd.merge(dg3, merged_small ,on="ident", how="inner")
    #Customising the required columns only
    merged_large=merged_large[['id', 'ident',"closed","frequency_mhz","iso_region","iso_country","description","type_x","type_y"]]
    merged_medium=merged_medium[['id', 'ident',"closed","frequency_mhz","iso_region","iso_country","description","type_x","type_y"]]
    merged_small=merged_small[['id', 'ident',"closed","frequency_mhz","iso_region","iso_country","description","type_x","type_y"]]
    #Displaying the Dataframe
    print(merged_large)
    print(merged_small)
    print(merged_medium)
  
    
    return merged_large , merged_small, merged_medium

merged=clean_csv()
#Buttin created in the main window for Cleaning CSV
cleanButton_CSV = tk.Button(text="      clean CSV     ", command=clean_csv, bg='green', fg='white',
                              font=('helvetica', 12, 'bold'))
canvas1.create_window(400, 130, window=cleanButton_CSV)

#Function to find the mean of large airports
def mean():
    global merged
    global df3
    global de3
    global dg3
    # Mean/Medium Mode of the Large Airport
    merged_large_mean=merged_large["frequency_mhz"].mean()
    merged_large_median=merged_large["frequency_mhz"].median()
    merged_large_mode=merged_large["frequency_mhz"].mode()
    #Display the values
    print(merged_large_mean)
    print(merged_large_median)
    print(merged_large_mode[0])
    #Separate window for display
    root1 = Tk()
    #Geometry Size
    root1.geometry("300x200")
    #Label Creation
    root1.title("Mean median/mode of large Airports")
    w = Label(root1, text= merged_large_mean, font = "50") 
    w = Label(root1, text= merged_large_median, font = "50") 
    w = Label(root1, text= merged_large_mode, font = "50") 
    w.pack()
    
#Button to display the mean/median /mode for large airports
meanButton_CSV = tk.Button(text="      Mean/median/mode of large airports     ", command=mean, bg='green', fg='white',
                              font=('helvetica', 12, 'bold'))
canvas1.create_window(700, 130, window=meanButton_CSV)

print("\n")

#Function to display small airports
def small_airport():
    global merged
    global df3
    global de3
    global dg3
    global merged_large
    global merged_small
    global merged_medium
    #Display Statement
    print("Small Airports")
    #Displaying the small airport window with selected columns
    merged_small=merged_small[['id', 'ident',"closed","frequency_mhz","iso_region","iso_country","description","type_x","type_y"]]
    print(merged_small)
    #Function calling to display small airport
    table3()
    
    

#Button to Display small Airorts
small_airportButton_CSV = tk.Button(text="      Small Airport Table   ", command=small_airport, bg='green', fg='white',
                              font=('helvetica', 12, 'bold'))
canvas1.create_window(150, 180, window=small_airportButton_CSV)
print("\n")

#Display the large Airport
def large_airport():
    global merged
    global df3
    global de3
    global dg3
    global merged_large
    global merged_small
    global merged_medium
    print("Large Airports")
    #Display the selected columns from the table
    merged_large=merged_large[['id', 'ident',"closed","frequency_mhz","iso_region","iso_country","description","type_x","type_y"]]
    print(merged_large)
    #Function calling to display large airport
    table1()
    return
    
    
#Button to create the large airports and fonting
large_airportButton_CSV = tk.Button(text="      Large Airport Table   ", command=large_airport, bg='green', fg='white',
                              font=('helvetica', 12, 'bold'))
canvas1.create_window(400, 180, window=large_airportButton_CSV)
print("\n")
#Function to display medium airports
def medium_airport():
    global merged
    global df3
    global de3
    global dg3
    global merged_large
    global merged_small
    global merged_medium
    
    print("Medium Airports")
    #Display the table with selected coulmns
    merged_medium=merged_medium[['id', 'ident',"closed","frequency_mhz","iso_region","iso_country","description","type_x","type_y"]]
    print(merged_medium)
    #Function to display medium airports
    table2()
#Button to create the medium airports
medium_airportButton_CSV = tk.Button(text="      Medium Airport Table   ", command=medium_airport, bg='green', fg='white',
                              font=('helvetica', 12, 'bold'))
canvas1.create_window(700, 180, window=medium_airportButton_CSV)

#Function to create mean for all the airports freq above 100 MHZ
def trans_mean():
    global merged
    global df3
    global de3
    global dg3
    #large airports
    merged_large_freq=merged_large[merged_large["frequency_mhz"]>100]
    merged_large_freq=merged_large_freq.mean()
    print("The mean of large airport greater than 100 MHZ is ",merged_large_freq)
    #Medium Airports
    merged_medium_freq=merged_medium[merged_medium["frequency_mhz"]>100]
    merged_medium_freq=merged_medium_freq.mean()
    print("The mean of medium airport greater than 100 MHZ is ",merged_medium_freq)
    #Small Airports
    merged_small_freq=merged_small[merged_small["frequency_mhz"]>100]
    merged_small_freq=merged_small_freq.mean()
    merged_freq=(merged_large_freq+ merged_small_freq+merged_medium_freq)/ 3
    
    
    print("The mean of small airport greater than 100 MHZ is \n",merged_small_freq)
    #Separate Window
    root1 = Tk()
    #Window size
    root1.geometry("300x200")
    root1.title("Mean median/mode of Airports above 100 MHZ")
    #Window Label
    w = Label(root1, text =merged_freq, font = "50") 
    
    w.pack()
    
#Button to display the mean median mode of all airports 
trans_meanButton_CSV = tk.Button(text="     Mean/median/mode of airports>100 MHZ   ", command=trans_mean , bg='green', fg='white',font=('helvetica', 12, 'bold'))
canvas1.create_window(200, 230, window=trans_meanButton_CSV)

#Function to create the plots- large airport
def plot_large():
    global merged
    global df3
    global de3
    global dg3
    global merged_large
    global merged_small
    global merged_medium
    #Grouping the count of frequency 
    merged_large.groupby('frequency_mhz').size()
    merged_large['freq_count'] = merged_large.groupby('frequency_mhz')['frequency_mhz'].transform('count')
    #Customising the figure size
    plt.figure(figsize=(15,5))
    #Customising the x limit
    plt.xlim(100,160,25)
    #Customising the Y limit
    plt.ylim(0,8,2.5)
    #Scatter Plot
    sns.scatterplot(  x=merged_large["frequency_mhz"], y=merged_large["freq_count"] ,color="red")
    #Title Creation
    plt.title("Scatterplot for large airports",color="Blue",fontsize="21")
    #Label for x axis
    plt.xlabel("Frequency", fontsize=16)
    #label for y label
    plt.ylabel("Count", fontsize=16)
    #legend 
    plt.legend("C")
    plt.show()
    
#button to display the large airport plots
plot_largeButton_CSV = tk.Button(text="     Plot Large Airports   ", command=plot_large, bg='green', fg='white',
                              font=('helvetica', 12, 'bold'))
canvas1.create_window(500, 230, window=plot_largeButton_CSV)
def plot_small():
    global merged
    global df3
    global de3
    global dg3
    global merged_large
    global merged_small
    global merged_medium
    #Groupng the count of frequency
    merged_small.groupby('frequency_mhz').size()
    merged_small['freq_count'] = merged_small.groupby('frequency_mhz')['frequency_mhz'].transform('count')
    #Customising the figure size
    plt.figure(figsize=(20,7))
    #Customising the x limit
    plt.xlim(100,400,25)
    #Customising the y limit
    plt.ylim(0,30,2.5)
    #Scatterplot
    sns.scatterplot(  x=merged_small["frequency_mhz"], y=merged_small["freq_count"] ,color="red")
    #Title creation
    plt.title("Scatterplot for small  airports",color="Blue",fontsize="21")
    #Label Creation
    plt.xlabel("Frequency", fontsize=16)
    plt.ylabel("Count", fontsize=16)

    plt.legend("P")
    plt.show()
    
#Button to display the small airport plots
plot_smallButton_CSV = tk.Button(text="     Plot small Airports   ", command=plot_small, bg='green', fg='white',
                              font=('helvetica', 12, 'bold'))
canvas1.create_window(700, 230, window=plot_smallButton_CSV)
def plot_medium():
    global merged
    global df3
    global de3
    global dg3
    global merged_large
    global merged_small
    global merged_medium
    merged_medium.groupby('frequency_mhz').size()
    merged_medium['freq_count'] = merged_medium.groupby('frequency_mhz')['frequency_mhz'].transform('count')
    plt.figure(figsize=(15,9))
    plt.xlim(100,400,25)
    plt.ylim(0,40,2.5)
    sns.scatterplot(  x=merged_medium["frequency_mhz"], y=merged_medium["freq_count"] ,color="red")

    plt.title("Scatterplot for medium airports",color="Blue",fontsize="21")
    plt.xlabel("Frequency", fontsize=16)
    plt.ylabel("Count", fontsize=16)

    plt.legend("single element")
    plt.show()
    
    
#Button to create the medium airports
plot_mediumButton_CSV = tk.Button(text="     Plot Medium Airports   ", command=plot_medium, bg='green', fg='white',
                              font=('helvetica', 12, 'bold'))
canvas1.create_window(200, 280, window=plot_mediumButton_CSV)
#Function to create the JSON Files
def storeJSON():
    global merged
    global df3
    global de3
    global dg3
    global merged_large
    global merged_small
    global merged_medium
    print('Saving to Json file')
    #Writing all the csv files to JSON 
    df1=merged_large.to_json("airport_large.json")
    de1=merged_small.to_json("airport_small.json")
    dg1=merged_medium.to_json("airport_medium.json")
    return df1,de1,dg1
    
#Lading the function 
load_json=storeJSON
#Button to create the 3 JSON Files
loadButton_CSV = tk.Button(text="      Store CSV as Json File     ", command=storeJSON, bg='green', fg='white',font=('helvetica', 12, 'bold'))
canvas1.create_window(600,280 ,window=loadButton_CSV)

#Function to restart the application
def restart():
    global merged
    global df3
    global de3
    global dg3
    global merged_large
    global merged_small
    global merged_medium
    global restart
    #Function repeats
    getCSV()
  
    
#Button to restart the application
restartButton_CSV = tk.Button(text="     Restart Application   ", command=restart, bg='green', fg='white',
                              font=('helvetica', 12, 'bold'))
canvas1.create_window(700, 380, window=restartButton_CSV)

# Existing the Application
def exitapplication():
    #Message box as the warning message
    MsgBox = tk.messagebox.askquestion('Exit Application', 'Are you sure you want to exit the application',
                                       icon='warning')
    if MsgBox == 'yes':
        root.destroy()

#Button to exit the application
exitButton = tk.Button(root, text='       Exit Application     ', command=exitapplication, bg='brown', fg='white',
                       font=('helvetica', 12, 'bold'))
canvas1.create_window(200, 380, window=exitButton)
#Infinte looping
root.mainloop()
```
